```python
from matplotlib.artist import Artist
def rec_gc(art, depth=0):
    if isinstance(art, Artist):
        print(" "*depth + str(art))
        for child in art.get_children():
            rec_gc(child, depth+2)
            
rec_gc(legend)
```

## line plots


```python
import numpy as np
linear_data = np.array([1,2,3,4,5,6,7,8])
quadratic_data = linear_data**2

plt.figure()
plt.plot(linear_data, '-o', quadratic_data, '-o')
```

```python
plt.plot([22,44,55], '--r')
```

```python
plt.xlabel('same data')
plt.ylabel('some other data')
plt.title('A title')
plt.legend(['Baseline', 'Competition', 'Us'])
```

```python
plt.gca().fill_between(range(len(linear_data)),
                      linear_data, quadratic_data,
                      facecolor='blue',
                      alpha=0.25)
```

## timeseries plotting


```python
import pandas as pd

plt.figure()
observation_dates= np.arange('2017-01-01', '2017-01-09', dtype= 'datetime64[D]')
observation_dates= list(map(pd.to_datetime, observation_dates))
plt.plot(observation_dates, linear_data, '-o',
        observation_dates, quadratic_data, '-o')
```

```python
x=plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
```

```python
plt.subplots_adjust(bottom=0.25)
```

```python
ax=plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')
ax.set_xlabel('Quadratic vs. Linear performance')
```

```python
ax.set_title("Quadratic ($x^2$) vs. Linear ($x$) performance")
```

## Barcharts

```python
plt.figure()
xvals = np.array([1,2,3,4,5,6,7,8])
plt.bar(xvals, linear_data, width= 0.4)
```

```python
new_xvals= []
for item in xvals:
    new_xvals.append(item+0.3)
    
    
plt.bar(new_xvals, quadratic_data, width=0.3, color='red')
```

```python
from random import randint
linear_err= [randint(0,15) for x in range(len(linear_data))]
plt.bar(xvals, linear_data, width=0.3, yerr=linear_err)
```

```python
plt.figure()
xvals= range(len(linear_data))
plt.bar(xvals, linear_data, width=0.3, color='b')
plt.bar(xvals, quadratic_data, width=0.3, bottom=linear_data, color='r')
```

```python
plt.figure()
xvals= range(len(linear_data))
plt.barh(xvals, linear_data, height=0.3, color='b')
plt.barh(xvals, quadratic_data, height=0.3, left=linear_data, color='r')
```

## dejunkyfy the plotting

```python
plt.figure()
languages=['Python','SQL','Java','C++','JavaScript']
pos=np.arange(len(languages))
popularity=[56,39,34,34,29]
plt.xticks(pos,languages, alpha=0.8)
bars=plt.bar(pos,popularity,align='center',linewidth=0,color='lightslategrey')
bars[0].set_color('#1F77B4')
#remove ylabels:
for bar in bars:
    plt.gca().text(bar.get_x()+bar.get_width()/2,bar.get_height()-5,
                  str(int(bar.get_height()))+'%',
    ha='center', color='w',fontsize=11)
plt.title('top 5 languages', alpha=0.8)

#removing ticks:
plt.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=True)   #to remove ticks

#removing axis:
for spine in plt.gca().spines.values():     
    spine.set_visible(False)
    
plt.show()
```
